```python
import random
import numpy as np
import matplotlib.pyplot as plt


```

```python

# Rate constants
c1 = 10  # Prey birth rate
c2 = 0.01  # Predator birth rate per prey
c3 = 10  # Predator death rate

# Initial conditions
Y1 = 1000  # Initial prey population
Y2 = 1000  # Initial predator population
t = 0  # Initial time

# Simulation parameters
total_time = 30  # Total simulation time
results = {"time": [], "Y1": [], "Y2": []}

for i in range(3):
    while t < total_time:
        # Calculate the propensities of the reactions
        a1 = c1 * Y1
        a2 = c2 * Y1 * Y2
        a3 = c3 * Y2

        # Calculate the total propensity
        a0 = a1 + a2 + a3
        if a0 == 0:
            results["time"].append(30)
            results["Y1"].append(Y1)
            results["Y2"].append(Y2)
            break

        # Generate two random numbers for choosing the next reaction time and type
        r1 = random.random()
        r2 = random.random()

        # Calculate the time until the next reaction
        tau = -np.log(r1) / a0

        # Determine which reaction occurs
        if r2 < a1 / a0:
            # Reaction 1: Prey reproduction
            Y1 += 1
        elif r2 < (a1 + a2) / a0:
            # Reaction 2: Predator-prey interaction
            Y1 -= 1
            Y2 += 1
        else:
            # Reaction 3: Predator death
            Y2 -= 1

        # Update the time
        t += tau

        # Record the current state
        results["time"].append(t)
        results["Y1"].append(Y1)
        results["Y2"].append(Y2)

    print("Results of simulation for initial condition, Y1=", 1000, "and Y2=",1000)
    #Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(results["time"], results["Y1"], label="Prey (Y1)")
    plt.plot(results["time"], results["Y2"], label="Predator (Y2)")
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.legend()
    plt.title("Lotka-Volterra Predator-Prey Simulation")
    plt.show()

    plt.plot(results["Y1"],results["Y2"])
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.title("Y1-Y2 phase space")
    plt.show()
```
